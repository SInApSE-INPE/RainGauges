```python
import os
import pandas as pd
import numpy as np

import geopandas as gpd

import cartopy.feature as cf
import cartopy.crs as ccrs

from matplotlib import patches
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# South America WMO project
SA_lon_min, SA_lon_max, SA_lat_min, SA_lat_max = -83,-33,-55,6
MX_lon_min, MX_lon_max, MX_lat_min, MX_lat_max = -120, -85, 12, 32
```

```python
SP = gpd.read_file(os.path.join('..','..','geometry','SaoPaulo.geojson'))
```

```python
# extract bounds of Sao Paulo geojson
# SP_lon_min, SP_lat_min, SP_lon_max, SP_lat_max = SP.total_bounds
```

```python
proj = ccrs.PlateCarree()
fig = plt.figure(figsize=(5,3),dpi=300)
gs = gridspec.GridSpec(1,2)

#==================================================================================================
ax1 = plt.subplot(gs[0, 0], projection = proj)
ax1.set_extent([-87,-30,-60,15],crs=proj)
ax1.add_feature(cf.COASTLINE,linewidth=0.1, linestyle='-',edgecolor='k')
ax1.add_feature(cf.BORDERS, linewidth=0.1, linestyle='-', edgecolor='k')

# ax1.add_geometries(SP.geometry, ccrs.PlateCarree(), edgecolor='k', facecolor='None', linewidth=0.4, alpha=0.5)

gl = ax1.gridlines(crs=proj,draw_labels=True,linewidth=0.2,color='gray',alpha=0.5,linestyle='--')
gl.top_labels = False
gl.bottom_labels = True
gl.left_labels = True
gl.right_labels = False
gl.xlabel_style = {'size': 6, 'color': 'k'}
gl.ylabel_style = {'size': 6, 'color': 'k'}

ax1.add_patch(patches.Rectangle((SA_lon_max,SA_lat_max),SA_lon_min-SA_lon_max,SA_lat_min-SA_lat_max,linestyle='--',linewidth=0.9,edgecolor='r',facecolor='None',zorder=5))

ax1.set_title(f'(a) South America', fontsize=5, loc='left')
ax1.set_title(f'{abs(SA_lon_min)}W to {abs(SA_lon_max)}W and {abs(SA_lat_min)}S to {abs(SA_lat_max)}N', fontsize=5, loc='right')

#==================================================================================================
ax1 = plt.subplot(gs[0, 1], projection = proj)
ax1.set_extent([-125, -80, 8, 35],crs=proj)
ax1.add_feature(cf.COASTLINE,linewidth=0.1, linestyle='-',edgecolor='k')
ax1.add_feature(cf.BORDERS, linewidth=0.1, linestyle='-', edgecolor='k')

# ax1.add_geometries(SP.geometry, ccrs.PlateCarree(), edgecolor='k', facecolor='None', linewidth=0.4, alpha=0.5)

gl = ax1.gridlines(crs=proj,draw_labels=True,linewidth=0.2,color='gray',alpha=0.5,linestyle='--')
gl.top_labels = False
gl.bottom_labels = True
gl.left_labels = True
gl.right_labels = False
gl.xlabel_style = {'size': 6, 'color': 'k'}
gl.ylabel_style = {'size': 6, 'color': 'k'}

ax1.add_patch(patches.Rectangle((MX_lon_max,MX_lat_max),MX_lon_min-MX_lon_max,MX_lat_min-MX_lat_max,linestyle='--',linewidth=0.9,edgecolor='r',facecolor='None',zorder=5))

ax1.set_title('(b) Mexico', fontsize=5, loc='left')
ax1.set_title(f'{abs(MX_lon_min)}W to {abs(MX_lon_max)}W and {abs(MX_lat_min)}S to {abs(MX_lat_max)}N', fontsize=5, loc='right')

#==================================================================================================
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05, wspace=0.2, hspace=0.1)

ax1.set_facecolor('white')
fig.patch.set_alpha(0)
```
