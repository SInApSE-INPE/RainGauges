```python
import numpy as np
import pandas as pd

import cartopy.crs as ccrs
import cartopy.feature as cf

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# South America
lon_min, lat_min, lon_max, lat_max = -85,-45,-32,13
```

```python
data_dir = '/home/arturo/Downloads/UCAR/2018/SURFACE_OBS:2018010212.csv'
DATA = pd.read_csv(data_dir, sep=',')
DATA.head(2)
```

```python
proj = ccrs.PlateCarree()
fig = plt.figure(figsize=(4,4),dpi=300)
gs = gridspec.GridSpec(1,1)

#==================================================================================================
ax1 = plt.subplot(gs[0, 0], projection = proj)
ax1.set_extent([lon_min, lon_max, lat_min, lat_max],crs=proj)
ax1.add_feature(cf.COASTLINE,linewidth=0.2, linestyle='-',edgecolor='k')
ax1.add_feature(cf.BORDERS, linewidth=0.2, linestyle='-', edgecolor='k')
ax1.scatter(DATA['Longitude'], DATA['Latitude'], s=1, c='k')
ax1.set_title(f'Total stations: {len(DATA)}', fontsize=8, loc='left')
ax1.set_title(f'2018', fontsize=8, loc='right')

ax1.set_facecolor('white')
fig.patch.set_alpha(0)
```
