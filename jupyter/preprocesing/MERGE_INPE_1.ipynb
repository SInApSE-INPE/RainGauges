```python
import os
import glob
import pandas as pd
import numpy as np
import xarray as xr

import cartopy.crs as ccrs
import cartopy.feature as cf

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
dir_base = os.path.join('/','media','arturo','Arturo','Data', 'Brazil', 'MERGE_INPE')
```

```python
file_list = glob.glob(os.path.join(dir_base, '2024', 'raw', '*.grib2'))
file_list = sorted(file_list)
file_len = len(file_list)
print(f'Number of files: {file_len}')
```

```python
dates_all = pd.date_range(
        start=f"2024-01-01 00:00:00",
        end=f"2024-12-31 23:59:00",
        freq="1h"
        ).strftime("%Y-%m-%d %H").tolist()
print(f'Number of dates: {len(dates_all)}')
```

```python
PRE = None
for nn in range(len(dates_all)):
    time_ref = pd.Timestamp(dates_all[nn]).strftime('%Y%m%d%H')

    try:
        tmp = xr.open_dataset(os.path.join(dir_base,'2024','raw',f'MERGE_CPTEC_{time_ref}.grib2'))
        if PRE is None:
            lat = tmp.latitude.values
            lon = tmp.longitude.values
            PRE = np.zeros((len(dates_all), len(lat), len(lon)))
        PRE[nn,:,:] = tmp.rdp.values

    except:
        print(f'FIle not found: {f'MERGE_CPTEC_{time_ref}.grib2'}')
        PRE[nn,:,:] = np.nan
```

```python
lat = tmp.latitude.values
lon = tmp.longitude.values
lon2d, lat2d = np.meshgrid(lon, lat)
```

```python
PRE_xr = xr.Dataset(data_vars={"PRE": (("time","lat","lon"), PRE)},
                    coords={'time': pd.to_datetime(dates_all), 'lat': lat, 'lon': lon-360},
                    attrs=dict(description='MERGE CPTEC hourly precipitation for 2024',))

PRE_xr.PRE.attrs["units"] = "mm/h"
PRE_xr.PRE.attrs["long_name"] = "precipitation"
PRE_xr.PRE.attrs["origname"] = "precipitation"

PRE_xr.lat.attrs["units"] = "degrees_north"
PRE_xr.lat.attrs["long_name"] = "Latitude"

PRE_xr.lon.attrs["units"] = "degrees_east"
PRE_xr.lon.attrs["long_name"] = "Longitude"
```

```python
yy_s = pd.to_datetime(dates_all[0]).year
mm_s = str(pd.to_datetime(dates_all[0]).month).zfill(2)
dd_s = str(pd.to_datetime(dates_all[0]).day).zfill(2)

yy_e = pd.to_datetime(dates_all[-1]).year
mm_e = str(pd.to_datetime(dates_all[-1]).month).zfill(2)
dd_e = str(pd.to_datetime(dates_all[-1]).day).zfill(2)

print(f'Start time: {dd_s}/{mm_s}/{yy_s}')
print(f'End time: {dd_e}/{mm_e}/{yy_e}')
```

```python
PRE_out = os.path.join(dir_base,f'MERGE-INPE_SA_1h_{yy_s}_{mm_s}_{dd_s}_{yy_e}_{mm_e}_{dd_e}.nc')
print(f'Export PRE data to {PRE_out}')
PRE_xr.to_netcdf(PRE_out)
```
