```python
import os
import subprocess
import pandas as pd
import concurrent.futures
```

```python
yy_s = 2008
yy_e = 2008
```

```python
# South America region
lat_min, lat_max, lon_min, lon_max = -50, 13, -85, -45
```

```python
dir_out = "/home/arturo/Downloads/UCAR"
```

```python
num_workers = 10
```

```python
dates_list = pd.date_range(start=f"{yy_s}-01-01 12:00:00", end=f"{yy_e}-12-10 12:00:00", freq="D").strftime("%Y-%m-%d %H:%M:%S").tolist()
print(f'Number of dates: {len(dates_list)}')
```

```python
def download_litte_r_for_date(date_str):
    filename = f'SURFACE_OBS:{date_str.split(" ")[0].replace("-", "")}12.csv'
    filename_dir = os.path.join(dir_out, filename)
    if os.path.isfile(filename_dir) and os.path.getsize(filename_dir) > 0:
        print(f"File {filename} already exists and is not empty. Skipping download.")
    else:
        print('Downloading data...')
        subprocess.run([
            "python",
            "../../python/download_litte_r.py",
            "-t", date_str,
            "-o", dir_out,
            "-lat_min", lat_min,
            "-lat_max", lat_max,
            "-lon_min", lon_min,
            "-lon_max", lon_max
        ])
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=num_workers) as executor:
    executor.map(download_litte_r_for_date, dates_list)
```
